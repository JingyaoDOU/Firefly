```python
import numpy as np
```

```python
with open('test.b','wb') as handle:
    handle.write(np.array(8*4).astype(np.int32)) ## size of the header
    handle.write(np.array(100).astype(np.int32)) ## number of particles
    handle.write(np.arange(300).reshape(-1,3).astype(np.float32)) ## coordinates
    handle.write(np.arange(300).reshape(-1,3).astype(np.float32)) ## velocities
    
```
